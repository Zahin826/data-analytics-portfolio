```python
import pandas as pd
```

```python
df = pd.read_csv('../data/olist_orders_dataset.csv')
print("Data loaded:", df.shape)
```

```python
print(df.dtypes)
```

```python
df['order_purchase_timestamp'] = pd.to_datetime(df['order_purchase_timestamp'])
```

```python
df['order_approved_at'] = pd.to_datetime(df['order_approved_at'])
```

```python
df['order_delivered_carrier_date'] = pd.to_datetime(df['order_delivered_carrier_date'])
```

```python
df['order_delivered_customer_date'] = pd.to_datetime(df['order_delivered_customer_date'])
```

```python
df['order_estimated_delivery_date'] = pd.to_datetime(df['order_estimated_delivery_date'])
```

```python
print(df.dtypes)
```

```python
print(df[df['order_approved_at'].isnull()]['order_status'].value_counts())
print("---")
print(df[df['order_delivered_customer_date'].isnull()]['order_status'].value_counts())
```

```python
# How many rows have order_status = 'delivered'?
print("Total orders:", len(df))

# Keep only delivered orders for our analysis
# We'll save the full dataset separately
df_delivered = df[df['order_status'] == 'delivered'].copy()
print("Delivered orders:", len(df_delivered))

# Check missing values in our clean dataset
print("\nMissing values in delivered orders:")
print(df_delivered.isnull().sum())
```

```python
# Drop rows with any remaining missing values
df_delivered = df_delivered.dropna()

# Verify
print("Final clean dataset:", df_delivered.shape)
print("Missing values:", df_delivered.isnull().sum().sum())
```

```python
# Calculate actual delivery time in days
df_delivered['delivery_days'] = (
    df_delivered['order_delivered_customer_date'] - 
    df_delivered['order_purchase_timestamp']
).dt.days

# Quick check
print(df_delivered['delivery_days'].describe())
```
